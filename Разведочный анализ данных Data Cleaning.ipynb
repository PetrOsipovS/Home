# 2. Выведите границы для нормальных значений года выпуска, используя интерквартильный размах. Воспользуйтесь материалами из видео. Посмотрите, входят ли минимальное и максимальное значения в интервал между полученными границами.

```python
def calculate_outliers(data):
    q25 = data.quantile(0.25)
    q75 = data.quantile(0.75)
    iqr = q75 - q25
    boundaries = (q25 - 1.5 * iqr, q75 + 1.5 * iqr)
    return boundaries
```

```python
boundaries = calculate_outliers(df_types.year)
boundaries
```

min 1915.000 значение не входит в интервал между полученными границами, max 2022.000 входит в интервал между полученными границами.

# 3. Посчитайте и выведите количество выбросов в годе выпуска автомобиля.

```python
is_outlier = (df_types.year < boundaries[0]) | (df_types.year > boundaries[1])
is_outlier
```
